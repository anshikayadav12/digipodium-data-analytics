```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rainfall= [17,10,10,10,26,59]
months = ['jan','feb','mar', 'apr','may','jun']
```

```python
plt.style.available
```

```python
plt.style.use('ggplot')#theme
```

```python
#graph chart
plt.bar(months,rainfall,color='#ff7702')
```

```python
#pie chart
plt.pie(
    rainfall,labels=months,autopct='%.1f%%',explode=(0,0,0,0,0,0.01)
)
plt.show()
```

```python
#donut
plt.pie(
    rainfall,labels=months,
    autopct='%.1f%%',
    wedgeprops={'width':0.4,'edgecolor':'#001100','linewidth':2},
    pctdistance=.8

)
plt.show()
```

```python
plt.boxplot(rainfall,
            vert=False,
            showmeans=True,
            meanline=True,
            meanprops={'color':'red'}
            )
plt.show()
```
