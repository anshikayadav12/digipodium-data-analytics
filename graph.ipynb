```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rainfall= [17,10,10,10,26,59]
months = ['jan','feb','mar', 'apr','may','jun']
```

```python
plt.style.available
```

```python
plt.style.use('ggplot')#theme
```

```python
#graph chart
plt.bar(months,rainfall,color='#ff7702')
```

```python
#pie chart
plt.pie(
    rainfall,labels=months,autopct='%.1f%%',explode=(0,0,0,0,0,0.01)
)
plt.show()
```

```python
#donut
plt.pie(
    rainfall,labels=months,
    autopct='%.1f%%',
    wedgeprops={'width':0.4,'edgecolor':'#001100','linewidth':2},
    pctdistance=.8

)
plt.show()
```

```python
plt.boxplot(rainfall,
            vert=False,
            showmeans=True,
            meanline=True,
            meanprops={'color':'red'}
            )
plt.show()
```

```python
t = np.arange(-10,10,.2)
y=np.sin(t)
fig,ax=plt.subplots(figsize=(5,3))
ax.plot(t,y,label='sin wave')
ax.legend()
ax.set_xlabel('time')
ax.set_ylabel('amplitude')
```

```python
t = np.arange(-10,10,.2)
y=np.sin(t)
fig,ax=plt.subplots(figsize=(5,3))
ax.plot(t,y,label='sin wave',
        color='purple',marker='o',linestyle=':',linewidth=2)
ax.legend()
ax.set_xlabel('time')
ax.set_ylabel('amplitude')
```

```python
t = np.arange(-10,10,.2)
y=np.sin(t)
fig,ax=plt.subplots(figsize=(5,3))
ax.plot(t,y,label='sin wave',
        color='green',marker='h',
        markersize=5,markerfacecolor='blue')
ax.legend()
ax.set_xlabel('time')
ax.set_ylabel('amplitude')
```

```python
t = np.arange(-10,10,.2)
y=np.cos(t)
fig,ax=plt.subplots(figsize=(5,3))
ax.plot(t,y,label='sin wave',
        color='green',marker='h',
        markersize=5,markerfacecolor='blue',
        alpha=.3)
ax.legend()
ax.set_xlabel('time')
ax.set_ylabel('amplitude')
```

```python
t = np.arange(-10,10,.2)
y=np.sin(t)
y2 = np.cos(t)
fig,ax=plt.subplots(figsize=(5,3))
ax.plot(t,y,label='sin wave',
        color='green',marker='h',
        markersize=5,markerfacecolor='blue',
        alpha=.5)
ax.plot(t,y2,label='cosine wave',
        color='brown',marker='h',
        markersize=5,markerfacecolor='brown',
        alpha=.7)
ax.legend()
ax.set_xlabel('time')
ax.set_ylabel('amplitude')
```
